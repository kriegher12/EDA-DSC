```python
Math_filtered = df.query('Math_Marks > 90')
Physics_filtered = df.query('Physics_Marks > 90')
Chemistry_filtered = df.query('Chemistry_Marks > 90')

#The above used query function allows me to query or question or ask answers from the dataset, herein I am asking the dataset to filter out the marks above 90.

Math_stat = Math_filtered.groupby('Gender')['Math_Marks'].count
Phy_stat = Physics_filtered.groupby('Gender')['Physics_Marks'].count
Chem_stat = Chemistry_filtered.groupby('Gender')['Chemistry_Marks'].count

#Now, I have created a variable storing the number of each subject, grouped by the gender

plt.figure(figsize=(14,6))

plt.subplot(1,3,1)
Phy_stat().plot(kind = 'pie',autopct = '%.0f%%') #autopct is used to label the parts in a pie chart
plt.title("Physics Marks")
plt.ylabel(" ")
plt.tight_layout()

plt.subplot(1,3,2)
Chem_stat().plot(kind = 'pie',autopct = '%.0f%%')
plt.title("Chemistry Marks")
plt.ylabel(" ")
plt.tight_layout()

plt.subplot(1,3,3)
Math_stat().plot(kind = 'pie',autopct = '%.0f%%')
plt.title("Maths Marks")
plt.ylabel(" ")
plt.tight_layout()
```
