```python
fig, (ax1,ax2) = plt.subplots(ncols = 2, figsize=(10,4))
gender = df['Gender'].value_counts()
gender.plot(kind = 'bar',ax=ax1,x='Gender',title = 'Barchart representation of gender')
ax1.set_xlabel("Gender")

explode = [0,0.1]
ax2.pie(gender,autopct = '%1.1f%%', explode=explode)
ax2.set_title('Piechart representation of gender')
ax2.legend(df['Gender'],loc='upper right')
```

There are slightly more female students than mail with the difference being of around 3%.

```python
fig, (ax1,ax2) = plt.subplots(ncols = 2, figsize=(10,4))
job = df['Has_Part_Time_Job'].value_counts()
job.plot(kind = 'bar',ax=ax1,x='Has_Part_Time_Job',title = 'Barchart representation of job')
ax1.set_xlabel("Has_Part_Time_Job")

explode = [0,0.1]
ax2.pie(job,autopct = '%1.1f%%', explode=explode)
ax2.set_title('Piechart representation of job')
ax2.legend(df['Has_Part_Time_Job'],loc='upper right')
```

The proportion of students with a job is almost equal to the proportion without one.

```python
Math_filtered = df.query('Math_Marks > 90')
Physics_filtered = df.query('Physics_Marks > 90')
Chemistry_filtered = df.query('Chemistry_Marks > 90')

#The above used query function allows me to query or question or ask answers from the dataset, herein I am asking the dataset to filter out the marks above 90.

Math_stat = Math_filtered.groupby('Gender')['Math_Marks'].count
Phy_stat = Physics_filtered.groupby('Gender')['Physics_Marks'].count
Chem_stat = Chemistry_filtered.groupby('Gender')['Chemistry_Marks'].count

#Now, I have created a variable storing the number of each subject, grouped by the gender

plt.figure(figsize=(14,6))

plt.subplot(1,3,1)
Phy_stat().plot(kind = 'pie',autopct = '%.0f%%') #autopct is used to label the parts in a pie chart
plt.title("Physics Marks")
plt.ylabel(" ")
plt.tight_layout()

plt.subplot(1,3,2)
Chem_stat().plot(kind = 'pie',autopct = '%.0f%%')
plt.title("Chemistry Marks")
plt.ylabel(" ")
plt.tight_layout()

plt.subplot(1,3,3)
Math_stat().plot(kind = 'pie',autopct = '%.0f%%')
plt.title("Maths Marks")
plt.ylabel(" ")
plt.tight_layout()
```

```python
Phy_Marks, Math_Marks,Chem_Marks = "Physics_Marks", "Math_Marks","Chemistry_Marks"
corr = df[Phy_Marks].corr(df[Math_Marks])
print("Correlation between ", Phy_Marks, " and ", Math_Marks, "is: ", round(corr, 2))
```

As we can see the correlation between Math and Physics marks is -0.02 which is almost negligible.

Similarly, here are the correlation between the other subject:

```python
corr = df[Phy_Marks].corr(df[Chem_Marks])
print("Correlation between ", Phy_Marks, " and ", Chem_Marks, "is: ", round(corr, 2))

print(" ")

corr = df[Math_Marks].corr(df[Chem_Marks])
print("Correlation between ", Math_Marks, " and ", Chem_Marks, "is: ", round(corr, 2))
```
